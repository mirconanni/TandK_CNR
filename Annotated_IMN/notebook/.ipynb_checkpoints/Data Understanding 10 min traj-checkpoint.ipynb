#### Same vehicle, this time we use the csv with the coordinates to try to plot them

```python
df_v1_coord = pd.read_csv("Traj10min/vehicle1_coord.csv") 
```

```python
df_v1_coord.head(5)
```

```python
def swapXY(traj):
    for i in traj:
        i[0], i[1] = i[1], i[0]
    return traj
```

```python
coords = []

for el in  df_v1_coord["trajcoord"]:
    y = json.loads(el)
    c = swapXY(y["coordinates"])
    coords.append(y["coordinates"])
```

```python
df_v1_coord["coord"] = coords
```

```python
df_v1_coord.head(10)
```

```python
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 'darkblue', 'darkgreen', 'cadetblue', 
          'darkpurple', 'pink', 'lightblue', 'lightgreen', 'gray', 'black', 'lightgray']

m = folium.Map(location=[40.8, 19.5], zoom_start=8.4)
for i in range(200):
    my_PolyLine=folium.PolyLine(locations=coords[i], weight=5, color=colors[i%18])
    m.add_child(my_PolyLine)
    
mc = MarkerCluster()
for i in range(200):
    marker = folium.Marker(location=coords[i][0], popup='Traj_1_Init', icon=folium.Icon(color=colors[i%18],icon='play'))
    mc.add_child(marker)
    
    marker = folium.Marker(location=coords[i][-1], popup='Traj_1_End', icon=folium.Icon(color=colors[i%18],icon='stop'))
    mc.add_child(marker)

m.add_child(mc)
m
```
